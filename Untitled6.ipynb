```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as m
p.set_option('display.max_column', None)
p.set_option('display.max_rows', None)
```

```python
t=p.read_csv("TSLA.csv")
```

```python
t["Date"]=p.to_datetime(t["Date"])
```

```python
p.set_option("display.float_format",lambda x:"%.2f"%x)
```

```python
t.head()
```

```python
t[(t.columns.dtypes=="float64")]
```

```python
k=t.columns!="Date"
k=t.columns[k]
k
```

```python
t.plot(x="Date",y=k,subplots=True,layout=(3,2),figsize=(15,8),rot=45);
```

```python
f,ax=m.subplots(len(k),figsize=(15,8))
for i,j in enumerate(k):
    s.distplot(t[j],ax=ax[i])
    f.tight_layout(pad=1.1)
```

```python
fo,axx=m.subplots(len(k),figsize=(15,10))
for i,j in enumerate(k):
    s.boxplot(t[j],ax=axx[i],orient="h")
    fo.tight_layout(pad=1.1)
```

```python
t["Year"]=p.DatetimeIndex(t['Date']).year
```

```python
t.head()
```

```python
f,axx=m.subplots(len(k),figsize=(15,25))

for i,j in enumerate(k):
    t.groupby("Year")[j].plot(ax=axx[i],legend=True)
    fo.tight_layout(pad=1.1)
```

```python
f,a=m.subplots(figsize=(15,5))
t.groupby("Year")["Volume"].sum().plot();
```

```python
t["Month"]=p.DatetimeIndex(t['Date']).month
```

```python
t.groupby("Month")["Volume"].sum().plot();
m.grid(True)
```

```python
kk=t.groupby("Year")["Volume"].sum()
kk
```
